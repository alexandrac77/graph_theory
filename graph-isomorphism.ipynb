## Graph Isomorphism Problem
***

The Graph Isomorphism (GI) problem is a computational problem which decides whether two graphs are isomorphic. A graph is isomorphic when there is a bijective, one-to-one correspondence from the nodes of the first graph to the nodes of the second graph, where edge connections are respected.  

A bijection is a map from set A to set B that is one to one (injective) and onto (surjective).
- Injections are distinct, meaning no two elements can map to the same element.
- Surjections map an element x to every element y; for every y, there is an x such that f(x) = y. 

Given graphs G and H, there is a bijection between the vertex sets of G and H $f: V(G)  V(H)$ such that any two vertices u and v of G are adjacent in G if and only if $f(u) f(u)$ and $f(v)f(v)$ are adjacent in H (adjacency and non-adjacency is preserved). [extracted from wikipedia]
This can be denoted by $G  H$. 

<b> Two graphs can look different but still be isomorphic. </b>

<br>

GI emerged in the 1970s as one of the few natural problems in the complexity class NP that could neither be classified as being NP-complete or solvable with an efficient algorithm.

The problem is still open and unresolved in the respect that no efficient algorithm has been found to solve it.

Lszl Babai, a Hungarian professor of computer science and mathematics at Chicago University, developed a quasipolynomial time algorithm that theoretically makes the GI problem solvable for any graph of order $n$.

### Graph Terminology

##### Undirected Graphs
According to Michael Sipser, an undirected graph is a set of points with lines connecting some of the points. The points are called nodes or vertices, and the lines are called edges.

The degree of a node can be determined by the number of edges at that particular node.
No more than one edge is allowed between any two nodes. An edge can be allowed from a node to itself (self-loop).

<br>

In a graph <b>G</b> that contains nodes <b>i</b> and <b>j</b>, the pair <b>(i, j)</b> represents the edge that connects the nodes i and j. In an undirected graph the order of the pair does not matter. A pair can be represented in set notation: { i , j }.

<br>

##### Paths
Paths are a sequence of nodes in a graph that are connected by edges. 

A path that does not repeat any nodes is known as a `simple path.`

A graph is `connected` if every two nodes have a path between them. 

If a path starts and ends with the exact same node, it is a `cycle`.  A cycle can be regarded as `simple` if it contains a minimum of three nodes and only repeats the starting and ending nodes.

A graph is a `tree` if it is connected and has no simple cycles.

<br>

##### Directed Graphs
As well as undirected graphs, there are directed graphs. 

Michael Sipser defines a directed graph as a graph that has arrows instead of lines. The number of arrows pointing from a particular node is the outdegree of that node, and the number of arrows pointing to a particular node is the indegree.

He states that a path in which all the arrows point in the same direction as its steps is called a directed path. A directed graph is strongly connected if a directed path connects every two nodes. 

Directed graphs can depict binary relations.

### Graph Data Structure

A graph data structure is made up of nodes or verticies that are connected to each other through edges. Data is stored in the nodes and relationships between the data is expressed through the edges.
They can represent real life relationships between entities.

<br>

Some operations that can be performed on graph data structures are:
- Adding or removing a vertex or edge from a graph.
- Checking if the graph contains a certain value.
- Finding the path from one vertex to another vertex.

<br>

The most common ways to represent graphs in data structures is with Adjacency Matrix and Adjaceny List.

<b> <u> ADJACENCY MATRIX </u> </b>
- 2D array with V x V vertices. Each row and each column represent a vertex.
- If the value of any element `a[i][j]` equals 1, there is a path between vertex i and vertex j.
- If the value of any element `a[i][j]` equals 0, there is no path between the vertices i and j.
- Searching for edges between two vertices is fast, however a larger amount of space is required for every possible link betwen the vertices.

<b> <u> ADJACENCY LIST </u> </b>
- A graph is represented as an array of linked lists.
- A vertex is represented by the index of the array.
- The element in the linked list represents the vertices that form an edge with the index vertex.
- The elements of the array `Ai` is a list. It contains all the vertices adjacent to vertex `i`.
- An adjacency list is storage efficient as only the value for the edges of a given node is stored.

# Determining If Two Graphs Are Isomorphic
***

<br>

Below we draw two graphs G1 and G2 respectively, using the networkx library.

We check whether the graphs are isomorphic by implementing the is_isomorphic function. This function returns true if the graphs are isomorphic, and false if they are not.


```python
import networkx as nx
G1 = nx.Graph()
G1.add_nodes_from(['A', 'B', 'C', 'D', 'E'])
G1.add_edges_from([('A', 'C'),('A','B'),('A','D'),('A','E')])
nx.draw(G1, with_labels=1, node_color="lightgreen")
```

```python
G2= nx.star_graph(7)
nx.draw(G2, with_labels=1, node_color="orange")
```

```python
# Return true if the graphs are isomorphic, or false if they are not isomorphic.
nx.is_isomorphic(G1,G2)
```

<b>The graphs G1 and G2 are NOT isomorphic.</b>

## Computational Complexity of Graph Isomorphism Problem
***

Determining the computational complexity of the graph isomorphism problem is unsolved. Polynomial or NP-Complete  (nondeterministic polynomial) algorithms for the problem have not yet been produced to clarify the problems status.
 
One approach to solve this type of problem is to use a polynomial algorithm to approximate the solution. The answer obtained in this way may not be optimal but will be reasonably close.

To check if two graphs, A and B are isomorphic, every possible way to match the nodes in graph A with the nodes in graph B can be considered. This is a simple approach for very small graphs, however for graphs with n nodes, the number of possible matchings becomes n! (n factorial). This makes a brute-force approach impractical, as, for a graph with 10 nodes, the possible number of matchings to check exceeds 3 million.

A polynomial algorithm would be far more efficient, as polynomials grow much slower compared to factorials. 

Computational problems can be classfied into classes of complexity, according to the amount of resources needed to solve them. An algorithm solved in polynomial time is within the class P. The class NP describes nondeterministic algorithms. 
Efficient algorithms for problems in the NP-Complete class have never been found.

Graph isomorphism lies between both the P and NP-Complete classes. 


The closest to solving the graph isomorphism problem is Lszl Babai, who proposed a quasipolynomial ( $exp (({log}^n)^{O(1)})$ ) time algorithm that can theoretically tell if graphs are isomorphic.
<i>n is the number of vertices and log(n) is some polynomial in log(n)</i>


A quasipolynomial algorithm's running time, for a graph containing n nodes, can be compared to a polynomial with n raised to a certain power. However the quasipolynomial's power is not a constant one, but grows slowly.

Babai's algorithm is based upon the work of Luks exponential algorithm, and contains four new principals:
a group-theoretic result, a group-theoretic divide-and-conquer algorithm, and two combinatorial partitioning algorithms.
